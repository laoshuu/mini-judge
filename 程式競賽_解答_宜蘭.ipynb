#### 第一題

日向翔陽正在為了接下來的比賽練習自己最不擅長的接發球，希望能每天記錄接發球成功率，已知接發球成功率公式如下: 接發球成功率(%)=好球÷ 接球總數×100，而接發球結果只有好球與壞球二種，請你幫日向計算今日的接發球成功率。

#### 說明：

- good是接球接得好的球數，bad則是接球接得壞的球數

#### 限制:

- 接發球結果只有好球與壞球二種

#### 範例 1


**input**: good = 50, bad = 50


**output**: 50

#### 範例 2


**input**: good = 120, bad = 30


**output**: 80

```python
def sol_1(good: int, bad: int) -> str:
    percentage = (good / (good + bad)) * 100
    return percentage
```

```python
student.submit(1, sol_1)
```

### 第三題

比賽的日子終於來臨，烏野高校和音駒高校正在進行排球比賽，排球比賽通常25分就能夠獲勝，但有特別的規則是若是兩隊同時24分，稱為"deuce"，必須要贏對方兩分才能夠獲勝。
給你烏野高校和音駒高校的分數，請你告訴我們誰獲得勝利，並且是否有發生deuce的情況。

#### 說明：

- wuya是烏野的分數，ing是音駒的分數

#### 限制:

- 一定會有一隊獲得勝利

#### 範例 1


**input**: wuya = 25, ing = 11


**output**: '烏野勝，沒有deuce'

#### 範例 2


**input**: wuya = 25, ing = 27


**output**: '音駒勝，有deuce'

```python
def sol_3(wuya: int, ing: int) -> str:
    if wuya > ing:
        if ing >= 24:
            return '烏野勝，有deuce'
        else:
            return '烏野勝，沒有deuce'
    else:
        if wuya >= 24:
            return '音駒勝，有deuce'
        else:
            return '音駒勝，沒有deuce'
```

影山去東京參加青年隊集訓，在隊內練習中，他的舉球成功率是85%，成功率越高越好，他想要知道他在集訓中排名第幾。<br>
給你其它選手的成績，請告訴影山他是第幾名。

#### 說明：

- scores 是其他人的舉球成功率
- 每個人的成績都不同
- 影山的成績是 85%

#### 限制:

- 100 ≥ scores  ≥ 0

#### Sample 1


**input**: scores = [50, 55, 60, 65, 70, 90]


**output**: 2


#### Sample 2


**input**: scores = [68, 72, 74, 88, 90]


**output**: 3

```python
def sol_4(scores: list) -> int:
    ranking = 1
    for score in scores:
        if score > 85:
            ranking = ranking + 1
    return ranking
```

影山集訓回來後變得更強了，能夠知道哪一個隊友狀態比較好再將球舉給那個隊友，並且每個隊友喜歡打的戰術名稱都不一樣。<br>
如果狀態好的人是澤村大地的話，就是舉「四號球」。
如果狀態好的人是東峰旭的話，就是舉「背後二號球」。
如果狀態好的人是日向翔陽的話，就是舉「怪人快攻」
給你接下來五球狀態好的人的名字，請你告訴影山接下來要舉什麼球吧！

#### 說明：

- players 是接下來五球狀態好的人的名字

#### Sample 1


**input**: players = ["日向翔陽", "東峰旭, "東峰旭", "日向翔陽", "日向翔陽"]


**output**: ["怪人快攻", "背後二號球", "背後二號球", "怪人快攻", "怪人快攻"]


#### Sample 2


**input**: players = ["澤村大地", "澤村大地", "日向翔陽", "東峰旭", "日向翔陽"]


**output**: ["四號球", "四號球", "怪人快攻", "背後二號球", "怪人快攻"]

```python
def sol_5(players: list) -> list:
    output = [0,0,0,0,0]
    for i in range(len(players)):
        if players[i] == "日向翔陽":
            output[i] = "怪人快攻"
        elif players[i] == "東峰旭":
            output[i] = "背後二號球"
        else:
            output[i] = "四號球"
    return output
```

在吃完好吃的糖葫蘆後，芙利蓮一行人來到魔法森林想透過消滅怪物來練習法術，已知裡面的怪物有四種，打敗四種怪物所需要的體力分別是[1, 5, 10, 50]，而擊敗四隻怪物所需的時間相同。芙利蓮一行人因爲想節省時間，不想打太多怪物，所以希望多擊敗所需體力較高的怪物來耗光體力。<br>
告訴你芙利蓮一行人的體力值，請告訴他們共打敗幾隻怪物。

#### 說明：

- 只需要輸出總共有幾隻怪物，不論種類
- power 是一行人的體力值
- 需要耗光全部體力

#### 範例 1
**Input:**

power = 200

**Output:**

4
#### 範例 2
**Input:**

power = 137

**Output:**

8

```python
def sol_7(power: int) -> int:
    monster = [50, 10, 5, 1]
    count = 0
    for hp in monster:
        while power - hp >= 0:
            power = power - hp
            count = count + 1
    return count
```

### 第八題

打敗森林裡的怪物們後，芙莉蓮一行人繼續前往洞穴探險，結果發現洞穴異常黑暗，好在洞口的樹上長滿了許多發光果，有了就可以照亮路程。為了環境永續，芙莉連一行人只能選擇兩棵樹拔下發光果。
我們知道整趟旅程中需要的發光果數量，以及每棵樹上有多少顆發光果，請告訴芙莉連一行人他們能不能夠湊到正確的發光果數量。


#### 說明：

- 一定要選兩棵樹摘發光果，不能只選一個
- fruits 是每棵樹發光果的數量
- target 是所需要的發光果數量
- 為了不浪費資源，發光果數量需要完全正確

#### 範例 1
**Input:**

fruits = [2, 4, 3, 8], target = 5

**Output:**

True
#### 範例 2
**Input:**

fruits = [1, 2, 4, 3, 5], target = 2

**Output:**

False

```python
def sol_8(fruits: list, target: int) -> bool:
    result = False
    for i in range(len(fruits) - 1):
        for j in range(i, len(fruits) - 1):
            if fruits[i] + fruits[j+1] == target:
                result = True
    return result
```

```python
student.run(8, sol_8)
```

```python
student.submit(8, sol_8)
```
